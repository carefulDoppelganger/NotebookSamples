```python
import numpy as np
import pandas as pd

import os
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.optimizers.legacy import Adam
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasRegressor
```

```python
data = pd.read_csv('train_scaled.csv')
test = pd.read_csv('test_scaled.csv')
```

```python
data.head()
```

```python
useddata = data[['News Sentiment', 'TickerName', 'close', 'volume']]
useddata.head()
```

```python
training_data = {}
training_label = {}
sequence_length = 30

for ticker in useddata["TickerName"]:
    
    temp_data = useddata.loc[useddata["TickerName"] == ticker, ['close', 'News Sentiment', 'volume']].values

    X, y = [], []

    for i in range(len(temp_data) - sequence_length):
        X.append(temp_data[i:i + sequence_length, :])
        y.append(temp_data[i + sequence_length, 0])

    X, y = np.array(X), np.array(y)
    
    training_data[ticker] = X
    training_label[ticker] = y
```

```python
def LSTM_model():
    model = Sequential()

    model.add(LSTM(units=100, return_sequences=True, input_shape=(30, 3)))
    model.add(Dropout(0.25))

    model.add(LSTM(units=100, return_sequences=True))
    model.add(Dropout(0.25))

    model.add(LSTM(units=100))
    model.add(Dropout(0.25))

    model.add(Dense(units=1))

    return model

model = LSTM_model()
model.summary
model.compile(
    optimizer=Adam(learning_rate=0.005), 
    loss='mean_squared_error'
)
```

```python
kscp = training_data["KSCP"]
kscp_lab = training_label["KSCP"]

checkpointer = ModelCheckpoint(
    filepath = 'best.hdf5', 
    verbose = False, 
    save_best_only = True
)

model.fit(kscp,
    kscp_lab, 
    epochs=50, 
    batch_size = 40,
    callbacks = [checkpointer]
)
```
